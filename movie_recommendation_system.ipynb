# Movie Recommendation System using Collaborative Filtering

## Data Science Internship Project

This project builds a movie recommendation system using collaborative filtering. The system recommends movies to users based on similar user ratings.


## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

## Step 2: Load the Dataset

Download the MovieLens dataset and place **ratings.csv** in the same folder.

```python
df = pd.read_csv('ratings.csv')
df.head()
```

## Step 3: Explore the Dataset

```python
print('Dataset Shape:', df.shape)
print('\nDataset Info:')
print(df.info())

print('\nSummary Statistics:')
print(df.describe())
```

## Step 4: Create User-Item Matrix

Rows = Users  
Columns = Movies  
Values = Ratings

```python
user_movie_matrix = df.pivot_table(
    index='user_id',
    columns='movie_id',
    values='rating'
).fillna(0)

user_movie_matrix.head()
```

## Step 5: Compute User Similarity Matrix

We use cosine similarity to measure similarity between users.

```python
user_similarity = cosine_similarity(user_movie_matrix)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)

user_similarity_df.head()
```

## Step 6: Recommendation Function

```python
def recommend_movies(user_id, top_n=5):
    
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:]

    weighted_scores = np.dot(similar_users, user_movie_matrix)

    recommendations = pd.Series(weighted_scores, index=user_movie_matrix.columns)

    return recommendations.sort_values(ascending=False).head(top_n)
```

## Step 7: Get Movie Recommendations

```python
recommend_movies(user_id=1)
```

## Step 8: Improved Recommendation (Unwatched Movies Only)

```python
def recommend_movies_filtered(user_id, top_n=5):

    watched_movies = user_movie_matrix.loc[user_id]
    unwatched_movies = watched_movies[watched_movies == 0].index

    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:]

    weighted_scores = np.dot(similar_users, user_movie_matrix)

    recommendations = pd.Series(weighted_scores, index=user_movie_matrix.columns)

    recommendations = recommendations[unwatched_movies]

    return recommendations.sort_values(ascending=False).head(top_n)
```

```python
recommend_movies_filtered(1)
```
